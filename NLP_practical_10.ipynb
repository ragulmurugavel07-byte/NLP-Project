```python
import re

patterns = {
    r'hi|hello|hey': ["Hello!", "Hey! How can I help you?"],
    r'how are you': ["I am fine!", "Doing great!", "I am here to help you."],
    r'what is your name': ["I am NLP chatbot.", "You can call me NLP Bot."],
    r'bye|exit': ["Goodbye!", "See you soon!", "Have a great day!"]
}

def chatbot_response(user_input):
  user_input = user_input.lower()
  for pattern_regex, response_message in patterns.items():
    if re.search(pattern_regex, user_input):
      return response_message
  return "I'm sorry, I didn't understand that."

print("Chatbot is running! Type 'bye' to exit.")

while True:
  user = input("You: ")
  response = chatbot_response(user)
  print("Bot: ", response)
  if re.search(r'bye|exit', user.lower()):
    break
```
